```python
df_copy=df.copy()
df_copy.head()
```

```python
x=df_copy.drop(columns=['price'], axis=1)
y=df_copy['price']
```

```python
print(x)
```

```python
df_copy.columns
```

```python
x_one_hot_encoding=pd.get_dummies(x,columns=['model', 'transmission', 'fuelType'],drop_first=True)
```

```python
x_one_hot_encoding=x_one_hot_encoding.astype(int)
```

```python
x_one_hot_encoding
```

```python
from sklearn.preprocessing import LabelEncoder

encode = LabelEncoder()
columns= ['model','transmission', 'fuelType' ]
xlable=x
for i in columns:
    xlable[i]=encode.fit_transform(xlable[i])
print(xlable)
```

```python
from sklearn.preprocessing import StandardScaler

numerical_cols=['year','mileage','tax','mpg','engineSize']
Scaler=StandardScaler()
x_one_hot_encoding[numerical_cols]=Scaler.fit_transform(x_one_hot_encoding[numerical_cols])
print(x_one_hot_encoding)
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize the scaler
scaler = StandardScaler()

numerical_cols = ['year', 'mileage', 'tax', 'mpg', 'engineSize'
xlable[numerical_cols] = scaler.fit_transform(xlable[numerical_cols])

print(xlable.head())
```

```python
xlable.columns
```

```python
xlable[['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg',
       'engineSize']]=scaler.fit_transform(xlable[['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg',
       'engineSize']])
```

```python
xlable.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_one_hot_encoding,y,test_size=0.3,random_state=42)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
print(y_pred)
```

```python
print(y_test)
```

```python
r2=r2_score(y_test,y_pred)
print(r2)
```

```python
print("Train R²:", model.score(x_train, y_train))
print("Test R²:", model.score(x_test, y_test))
```

```python
n=x_test.shape[0]
p=x_test.shape[1]
adjusted_r2=1-((1-r2)*(n-1))/(n-p-1)
print("Adjusted R^2 Score:",adjusted_r2)
```
